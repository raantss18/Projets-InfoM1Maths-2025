````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import warnings
warnings.filterwarnings('ignore')

# Biblioth√®ques pour le pr√©traitement
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Biblioth√®ques pour la mod√©lisation
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                           confusion_matrix, classification_report, roc_auc_score, roc_curve)

# Mod√®les de classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

# Pour l'analyse des erreurs
from sklearn.metrics import ConfusionMatrixDisplay
import scipy.stats as stats

# Configuration de l'affichage
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
rcParams['figure.figsize'] = (12, 8)
rcParams['font.size'] = 12

# %% [markdown]
# ## √âtape 2 : Chargement des donn√©es R√âELLES (SANS DONN√âES AL√âATOIRES)

# %%
# Charger VOS donn√©es r√©elles
file_path = 'Student Mental Health Analysis During Online Learning.csv'

try:
    df = pd.read_csv(file_path, encoding='utf-8')
    print("‚úÖ Donn√©es charg√©es avec succ√®s !")
    print(f"Dimensions du dataset : {df.shape}")
    print(f"\nAper√ßu des donn√©es (premi√®res 5 lignes) :")
    print(df.head())
    print(f"\nInformations sur les donn√©es :")
    print(df.info())
    print(f"\nStatistiques descriptives :")
    print(df.describe(include='all'))

    # V√©rifier les noms de colonnes
    print(f"\nüîç Noms des colonnes :")
    for i, col in enumerate(df.columns, 1):
        print(f"{i}. {col}")

    # V√©rifier les valeurs manquantes
    print(f"\nüßê Valeurs manquantes par colonne :")
    missing_values = df.isnull().sum()
    for col, count in missing_values.items():
        if count > 0:
            print(f"  {col}: {count} valeurs manquantes ({count/len(df)*100:.1f}%)")
    if missing_values.sum() == 0:
        print("  Aucune valeur manquante trouv√©e !")

except FileNotFoundError:
    print(f"‚ùå ERREUR CRITIQUE : Fichier non trouv√© √† l'emplacement : {file_path}")
    print("\n" + "="*60)
    print("ACTION REQUISE :")
    print("="*60)
    print("\nVeuillez v√©rifier et corriger le chemin du fichier :")
    print(f"Chemin actuel : {file_path}")
    print("\nOptions :")
    print("1. V√©rifiez que le fichier existe √† cet emplacement")
    print("2. V√©rifiez l'orthographe du nom du fichier")
    print("3. Essayez d'autres encodages (latin-1, iso-8859-1, cp1252)")
    print("\nExemples de chemins corrects :")
    print("  - Windows: 'C:/Users/VotreNom/Documents/Student_Mental_Health.csv'")
    print("  - Mac/Linux: '/home/utilisateur/Documents/Student_Mental_Health.csv'")
    print("\nPour t√©l√©charger des donn√©es r√©elles :")
    print("1. Visiter Kaggle : https://www.kaggle.com/datasets")
    print("2. Rechercher 'student mental health dataset'")
    print("3. T√©l√©charger et mettre √† jour le chemin dans le code")

    # Arr√™ter l'ex√©cution
    raise SystemExit("Arr√™t du programme. Veuillez corriger le chemin du fichier et r√©ex√©cuter.")

except UnicodeDecodeError:
    print(f"‚ùå ERREUR D'ENCODAGE : Probl√®me avec l'encodage du fichier")
    print("\nEssayez avec un encodage diff√©rent :")
    print("1. latin-1 : pour les fichiers europ√©ens")
    print("2. iso-8859-1 : similaire √† latin-1")
    print("3. cp1252 : pour Windows")
    print("\nExemple : df = pd.read_csv(file_path, encoding='latin-1')")
    raise SystemExit("Arr√™t du programme. Corrigez l'encodage et r√©ex√©cutez.")

except Exception as e:
    print(f"‚ùå Erreur inattendue : {e}")
    print("\nType d'erreur :", type(e).__name__)
    raise SystemExit(f"Arr√™t du programme. Erreur : {e}")

# %% [markdown]
# ## √âtape 3 : Nettoyage et pr√©paration des donn√©es

# %%
print("=== NETTOYAGE DES DONN√âES ===")

# Copie du dataframe original
df_clean = df.copy()

# 1. Supprimer les colonnes non n√©cessaires (comme 'Name' pour l'anonymat)
if 'Name' in df_clean.columns:
    print(f"Suppression de la colonne 'Name' (non n√©cessaire pour l'analyse)")
    df_clean = df_clean.drop('Name', axis=1)

# 2. V√©rifier et corriger les types de donn√©es
print("\nüîß Correction des types de donn√©es :")

# Colonnes qui devraient √™tre num√©riques
numeric_cols = ['Age', 'Screen Time (hrs/day)', 'Sleep Duration (hrs)', 'Physical Activity (hrs/week)']
for col in numeric_cols:
    if col in df_clean.columns:
        # Essayer de convertir en num√©rique
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
        print(f"  {col}: converti en num√©rique")

# 3. G√©rer les valeurs manquantes
print("\nüßπ Gestion des valeurs manquantes :")
missing_before = df_clean.isnull().sum().sum()

for col in df_clean.columns:
    if df_clean[col].isnull().sum() > 0:
        if df_clean[col].dtype in ['int64', 'float64']:
            # Pour les num√©riques : remplacer par la m√©diane
            df_clean[col] = df_clean[col].fillna(df_clean[col].median())
            method = "m√©diane"
        else:
            # Pour les cat√©gorielles : remplacer par le mode
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown')
            method = "mode"
        print(f"  {col}: {df_clean[col].isnull().sum()} valeurs manquantes remplac√©es par {method}")

missing_after = df_clean.isnull().sum().sum()
print(f"\n‚úÖ Valeurs manquantes totales : {missing_before} ‚Üí {missing_after}")

# 4. Standardiser les valeurs cat√©gorielles
print("\nüìù Standardisation des valeurs cat√©gorielles :")

# Nettoyer les cha√Ænes de caract√®res
categorical_cols = df_clean.select_dtypes(include=['object']).columns
for col in categorical_cols:
    # Supprimer les espaces en d√©but/fin
    df_clean[col] = df_clean[col].astype(str).str.strip()
    # Capitaliser la premi√®re lettre
    df_clean[col] = df_clean[col].str.title()
    print(f"  {col}: valeurs standardis√©es")

# 5. V√©rifier les valeurs uniques
print("\nüîç Valeurs uniques par colonne cat√©gorielle :")
for col in categorical_cols:
    unique_vals = df_clean[col].unique()
    print(f"  {col}: {len(unique_vals)} valeurs uniques")
    if len(unique_vals) < 10:  # Afficher seulement si peu de valeurs
        print(f"    ‚Üí {sorted(unique_vals)}")

# 6. Afficher un aper√ßu apr√®s nettoyage
print("\nüìä Aper√ßu apr√®s nettoyage :")
print(df_clean.head())
print(f"\nShape final : {df_clean.shape}")

# %% [markdown]
# ## √âtape 4 : Analyse Exploratoire des Donn√©es (EDA)

# %%
print("=== ANALYSE EXPLORATOIRE DES DONN√âES ===")

# 1. Distribution des variables cat√©gorielles
categorical_cols = df_clean.select_dtypes(include=['object']).columns.tolist()

if categorical_cols:
    n_cats = len(categorical_cols)
    n_rows = (n_cats + 2) // 3  # Arrondir vers le haut

    fig, axes = plt.subplots(n_rows, 3, figsize=(15, 4*n_rows))
    if n_cats > 1:
        axes = axes.flatten()
    else:
        axes = [axes]

    for idx, col in enumerate(categorical_cols):
        if idx < len(axes):
            value_counts = df_clean[col].value_counts()

            # Cr√©er un gradient de couleurs sans seaborn
            colors = plt.cm.Set3(np.linspace(0, 1, len(value_counts)))

            bars = axes[idx].bar(value_counts.index, value_counts.values, color=colors, edgecolor='black')
            axes[idx].set_title(f'Distribution de {col}', fontsize=12, fontweight='bold')
            axes[idx].tick_params(axis='x', rotation=45)
            axes[idx].set_ylabel('Nombre')
            axes[idx].grid(axis='y', alpha=0.3)

            # Ajouter les nombres sur les barres
            for bar in bars:
                height = bar.get_height()
                axes[idx].text(bar.get_x() + bar.get_width()/2., height,
                             f'{int(height)}', ha='center', va='bottom')

    # Cacher les axes non utilis√©s
    for idx in range(len(categorical_cols), len(axes)):
        axes[idx].set_visible(False)

    plt.tight_layout()
    plt.show()

# 2. Distribution des variables num√©riques
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()

if numeric_cols:
    n_nums = len(numeric_cols)
    n_rows = (n_nums + 1) // 2

    fig, axes = plt.subplots(n_rows, 2, figsize=(15, 4*n_rows))
    if n_nums > 1:
        axes = axes.flatten()
    else:
        axes = [axes]

    for idx, col in enumerate(numeric_cols):
        if idx < len(axes):
            # Histogramme
            n_bins = min(30, len(df_clean[col].unique()))
            axes[idx].hist(df_clean[col], bins=n_bins, edgecolor='black', alpha=0.7, color='skyblue')
            axes[idx].set_title(f'Distribution de {col}', fontsize=12, fontweight='bold')
            axes[idx].set_xlabel(col)
            axes[idx].set_ylabel('Fr√©quence')
            axes[idx].grid(alpha=0.3)

            # Ajouter des lignes pour moyenne et m√©diane
            mean_val = df_clean[col].mean()
            median_val = df_clean[col].median()

            axes[idx].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Moyenne: {mean_val:.2f}')
            axes[idx].axvline(median_val, color='green', linestyle='-.', linewidth=2, label=f'M√©diane: {median_val:.2f}')
            axes[idx].legend()

    # Cacher les axes non utilis√©s
    for idx in range(len(numeric_cols), len(axes)):
        axes[idx].set_visible(False)

    plt.tight_layout()
    plt.show()

# 3. Boxplots pour d√©tecter les outliers
if numeric_cols:
    fig, axes = plt.subplots(1, min(len(numeric_cols), 4), figsize=(15, 6))
    if len(numeric_cols) > 1:
        axes = axes.flatten()
    else:
        axes = [axes]

    for idx, col in enumerate(numeric_cols[:4]):  # Limiter √† 4 boxplots max
        if idx < len(axes):
            box = axes[idx].boxplot(df_clean[col], patch_artist=True)
            box['boxes'][0].set_facecolor('lightblue')
            box['boxes'][0].set_edgecolor('black')
            axes[idx].set_title(f'Boxplot de {col}', fontsize=12, fontweight='bold')
            axes[idx].set_ylabel(col)
            axes[idx].grid(axis='y', alpha=0.3)

            # Calculer et afficher les outliers
            Q1 = df_clean[col].quantile(0.25)
            Q3 = df_clean[col].quantile(0.75)
            IQR = Q3 - Q1
            outliers = df_clean[(df_clean[col] < Q1 - 1.5*IQR) | (df_clean[col] > Q3 + 1.5*IQR)][col]

            if len(outliers) > 0:
                axes[idx].text(0.95, 0.95, f'Outliers: {len(outliers)}',
                             transform=axes[idx].transAxes, ha='right', va='top',
                             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.5))

    plt.tight_layout()
    plt.show()

# %% [markdown]
# ## √âtape 5 : Analyse des corr√©lations et relations (SANS SEABORN)

# %%
print("=== ANALYSE DES CORR√âLATIONS ===")

# 1. Encodage temporaire pour la matrice de corr√©lation
df_encoded = df_clean.copy()
label_encoders = {}

for col in df_encoded.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = le

# 2. Matrice de corr√©lation SANS SEABORN
plt.figure(figsize=(12, 10))
correlation_matrix = df_encoded.corr()

# Cr√©er une heatmap avec matplotlib seulement
fig, ax = plt.subplots(figsize=(12, 10))
im = ax.imshow(correlation_matrix, cmap='coolwarm', aspect='auto')

# Ajouter les annotations
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        text = ax.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}',
                      ha='center', va='center', color='black', fontsize=9)

# Configurer les ticks
ax.set_xticks(np.arange(len(correlation_matrix.columns)))
ax.set_yticks(np.arange(len(correlation_matrix.index)))
ax.set_xticklabels(correlation_matrix.columns, rotation=45, ha='right')
ax.set_yticklabels(correlation_matrix.index)

# Titre et barre de couleur
ax.set_title('Matrice de Corr√©lation (valeurs encod√©es)', fontsize=16, fontweight='bold')
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel('Corr√©lation', rotation=-90, va='bottom')

plt.tight_layout()
plt.show()

# 3. Corr√©lations les plus fortes avec les variables cibles
print("\nüîó Corr√©lations avec les variables cibles potentielles :")

# Identifier les variables cibles potentielles (cat√©gorielles avec peu de valeurs)
potential_targets = []
for col in categorical_cols:
    if 2 <= len(df_clean[col].unique()) <= 5:
        potential_targets.append(col)

for target in potential_targets:
    print(f"\nüìà Corr√©lations avec '{target}':")
    target_encoded = label_encoders[target].transform(df_clean[target].astype(str))

    correlations = {}
    for col in df_encoded.columns:
        if col != target:
            corr = np.corrcoef(df_encoded[col], target_encoded)[0, 1]
            correlations[col] = abs(corr)

    # Trier par corr√©lation absolue
    sorted_corrs = sorted(correlations.items(), key=lambda x: x[1], reverse=True)[:5]

    for feature, corr in sorted_corrs:
        print(f"  {feature}: {corr:.3f}")

# 4. Relations sp√©cifiques entre variables cl√©s
print("\nüîç Relations cl√©s √† explorer :")

# Si nous avons les colonnes attendues
if 'Stress Level' in df_clean.columns:
    stress_levels = df_clean['Stress Level'].unique()
    print(f"\nüìä Stress Level par rapport aux autres variables :")

    # Variables √† analyser
    key_vars = ['Screen Time (hrs/day)', 'Sleep Duration (hrs)',
                'Physical Activity (hrs/week)', 'Age']

    # Filtrer les variables qui existent
    key_vars = [var for var in key_vars if var in df_clean.columns]

    if key_vars:
        n_vars = len(key_vars)
        n_rows = (n_vars + 1) // 2

        fig, axes = plt.subplots(n_rows, 2, figsize=(15, 4*n_rows))
        if n_vars > 1:
            axes = axes.flatten()
        else:
            axes = [axes]

        for idx, var in enumerate(key_vars):
            if idx < len(axes):
                data_to_plot = []
                labels = []

                for level in sorted(stress_levels):
                    subset = df_clean[df_clean['Stress Level'] == level]
                    if not subset.empty and var in subset.columns:
                        data_to_plot.append(subset[var].dropna())
                        labels.append(level)

                if data_to_plot:
                    bp = axes[idx].boxplot(data_to_plot, labels=labels, patch_artist=True)
                    # Colorier les boxplots
                    colors = ['lightgreen', 'lightblue', 'lightcoral', 'lightyellow'][:len(data_to_plot)]
                    for patch, color in zip(bp['boxes'], colors):
                        patch.set_facecolor(color)
                        patch.set_edgecolor('black')

                    axes[idx].set_title(f'{var} par Stress Level', fontsize=12, fontweight='bold')
                    axes[idx].set_ylabel(var)
                    axes[idx].grid(True, alpha=0.3)

        # Cacher les axes non utilis√©s
        for idx in range(len(key_vars), len(axes)):
            axes[idx].set_visible(False)

        plt.tight_layout()
        plt.show()

# %% [markdown]
# ## √âtape 6 : Pr√©paration des donn√©es pour la mod√©lisation

# %%
print("=== PR√âPARATION POUR LA MOD√âLISATION ===")

# 1. D√©finir les variables cibles
print("üéØ S√©lection des variables cibles :")

# D√©finition explicite des noms de colonnes cibles
STRESS_TARGET = 'Stress Level'
ACADEMIC_TARGET = 'Academic Performance Change' # Correction du nom de la colonne

# V√©rifier que ces colonnes existent avant de proc√©der
for col in [STRESS_TARGET, ACADEMIC_TARGET]:
    if col not in df_clean.columns:
        raise ValueError(f"‚ùå La colonne cible '{col}' n'existe pas dans le dataset. V√©rifie les noms exacts des colonnes.")

print(f"\n‚úÖ Variables cibles s√©lectionn√©es :")
print(f"  - T√¢che 1 (Stress/Principal): {STRESS_TARGET}")
print(f"  - T√¢che 2 (Performance acad√©mique/Secondary): {ACADEMIC_TARGET}")

# 2. Fonction de pr√©traitement
def prepare_data_for_modeling(df, target_column, test_size=0.2, val_size=0.25):
    """
    Pr√©pare les donn√©es pour la mod√©lisation avec s√©paration train/val/test
    et encode la variable cible num√©riquement.
    """
    print(f"\nüîß Pr√©paration des donn√©es pour '{target_column}':")

    # V√©rifier que la target existe
    if target_column not in df.columns:
        print(f"‚ùå Erreur: Colonne '{target_column}' non trouv√©e!")
        print(f"   Colonnes disponibles: {list(df.columns)}")
        raise ValueError(f"La colonne cible '{target_column}' n'existe pas dans le dataset.")


    # S√©parer features et target
    X = df.drop(target_column, axis=1)
    y = df[target_column]

    # Encoder la variable cible num√©riquement
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    target_classes = le.classes_ # Pour r√©f√©rence si n√©cessaire

    print(f"  Features: {X.shape[1]} colonnes")
    print(f"  Target: {len(target_classes)} classes ({', '.join(map(str, target_classes))})")

    # Identifier les types de colonnes
    numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()

    print(f"  Colonnes num√©riques: {len(numeric_features)}")
    print(f"  Colonnes cat√©gorielles: {len(categorical_features)}")

    # V√©rifier qu'il y a des features
    if len(numeric_features) + len(categorical_features) == 0:
        print("‚ùå Erreur: Aucune feature trouv√©e!")
        raise ValueError("Aucune feature trouv√©e!")

    # S√©paration 60-20-20
    # D'abord s√©parer test (20%)
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y_encoded, test_size=test_size, random_state=42, stratify=y_encoded
    )

    # Puis s√©parer train/val (75% train, 25% val du reste = 20% total)
    val_size_adjusted = val_size / (1 - test_size)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=val_size_adjusted, random_state=42, stratify=y_temp
    )

    print(f"\nüìä R√©partition des donn√©es :")
    print(f"  Train: {len(X_train)} √©chantillons ({len(X_train)/len(X)*100:.1f}%)")
    print(f"  Validation: {len(X_val)} √©chantillons ({len(X_val)/len(X)*100:.1f}%)")
    print(f"  Test: {len(X_test)} √©chantillons ({len(X_test)/len(X)*100:.1f}%)")

    # Cr√©er le pr√©processeur
    numeric_transformer = StandardScaler()
    categorical_transformer = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ])

    return X_train, X_val, X_test, y_train, y_val, y_test, preprocessor, le # Return LabelEncoder too

# 3. Pr√©parer les donn√©es pour les deux t√¢ches
print("\n" + "="*60)
print("PR√âPARATION DES DONN√âES POUR LES DEUX T√ÇCHES")
print("="*60)

# T√¢che 1 (Stress)
X_train_stress, X_val_stress, X_test_stress, y_train_stress, y_val_stress, y_test_stress, preprocessor_stress, le_stress = prepare_data_for_modeling(df_clean, STRESS_TARGET)

# T√¢che 2 (Performance acad√©mique)
X_train_academic, X_val_academic, X_test_academic, y_train_academic, y_val_academic, y_test_academic, preprocessor_academic, le_academic = prepare_data_for_modeling(df_clean, ACADEMIC_TARGET)

# %% [markdown]
# ## √âtape 7 : Construction et √©valuation des 8 mod√®les

# %%
def build_and_evaluate_models(X_train, X_val, y_train, y_val, preprocessor, task_name, label_encoder):
    """
    Construit et √©value 8 mod√®les de classification
    """
    print(f"\n{'='*60}")
    print(f"CONSTRUCTION DES MOD√àLES - {task_name}")
    print(f"{'='*60}")

    # Appliquer le pr√©traitement
    print("üîß Application du pr√©traitement...")
    try:
        X_train_processed = preprocessor.fit_transform(X_train)
        X_val_processed = preprocessor.transform(X_val)
    except Exception as e:
        print(f"‚ùå Erreur lors du pr√©traitement: {e}")
        return {}, None, None

    # D√©finir les 8 mod√®les
    models = {
        'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
        'Decision Tree': DecisionTreeClassifier(random_state=42),
        'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
        'Gradient Boosting': GradientBoostingClassifier(random_state=42, n_estimators=100),
        'SVM': SVC(probability=True, random_state=42),
        'K-Nearest Neighbors': KNeighborsClassifier(),
        'Naive Bayes': GaussianNB(),
        'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
    }

    results = {}

    for name, model in models.items():
        print(f"\nüèóÔ∏è  Entra√Ænement: {name}...")

        try:
            # Entra√Ænement
            model.fit(X_train_processed, y_train)

            # Pr√©dictions
            y_pred = model.predict(X_val_processed)

            # Calcul des m√©triques
            accuracy = accuracy_score(y_val, y_pred)
            precision = precision_score(y_val, y_pred, average='weighted', zero_division=0)
            recall = recall_score(y_val, y_pred, average='weighted', zero_division=0)
            f1 = f1_score(y_val, y_pred, average='weighted', zero_division=0)

            # AUC-ROC (si le mod√®le supporte predict_proba)
            if hasattr(model, "predict_proba"):
                y_pred_proba = model.predict_proba(X_val_processed)
                if len(label_encoder.classes_) == 2:
                    auc = roc_auc_score(y_val, y_pred_proba[:, 1])
                else:
                    auc = roc_auc_score(y_val, y_pred_proba, multi_class='ovr', average='weighted')
            else:
                auc = None

            # Stocker les r√©sultats
            results[name] = {
                'model': model,
                'accuracy': accuracy,
                'precision': precision,
                'recall': recall,
                'f1': f1,
                'auc': auc,
                'y_pred': y_pred
            }

            print(f"  ‚úÖ Accuracy: {accuracy:.4f}")
            print(f"  ‚úÖ F1-Score: {f1:.4f}")
            if auc:
                print(f"  ‚úÖ AUC-ROC: {auc:.4f}")

        except Exception as e:
            print(f"  ‚ùå Erreur avec {name}: {str(e)[:100]}...")
            continue

    return results, X_train_processed, X_val_processed

# Construire les mod√®les pour la t√¢che 1
print("\nüìä T√ÇCHE 1 : MOD√âLISATION DU STRESS")
stress_results, X_train_stress_processed, X_val_stress_processed = build_and_evaluate_models(
    X_train_stress, X_val_stress, y_train_stress, y_val_stress,
    preprocessor_stress, f"Pr√©diction de '{STRESS_TARGET}'", le_stress
)

# Construire les mod√®les pour la t√¢che 2 (Performance acad√©mique)
print("\nüìä T√ÇCHE 2 : MOD√âLISATION DE LA PERFORMANCE ACAD√âMIQUE")
academic_results, X_train_academic_processed, X_val_academic_processed = build_and_evaluate_models(
    X_train_academic, X_val_academic, y_train_academic, y_val_academic,
    preprocessor_academic, f"Pr√©diction de '{ACADEMIC_TARGET}'", le_academic
)

# Visualisation des performances SANS SEABORN
def plot_model_performance(results_dict, task_name):
    """
    Visualise les performances des mod√®les avec matplotlib uniquement
    """
    if not results_dict:
        print(f"‚ùå Aucun r√©sultat √† afficher pour {task_name}")
        return

    models = list(results_dict.keys())
    metrics = ['accuracy', 'f1', 'precision', 'recall']

    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.flatten()

    for idx, metric in enumerate(metrics):
        values = [results_dict[model][metric] for model in models]

        # Cr√©er un gradient de couleurs avec matplotlib
        colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(models)))

        bars = axes[idx].bar(models, values, color=colors, edgecolor='black')
        axes[idx].set_title(f'{metric.upper()} - {task_name}', fontsize=12, fontweight='bold')
        axes[idx].set_ylabel(metric.upper())
        axes[idx].tick_params(axis='x', rotation=45)
        axes[idx].set_ylim([0, 1])
        axes[idx].grid(axis='y', alpha=0.3)

        # Ajouter les valeurs sur les barres
        for bar, val in zip(bars, values):
            height = bar.get_height()
            axes[idx].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                         f'{val:.3f}', ha='center', va='bottom', fontsize=9)

    plt.suptitle(f'Performance des mod√®les - {task_name}', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Visualiser pour les deux t√¢ches
if stress_results:
    plot_model_performance(stress_results, f"Pr√©diction de '{STRESS_TARGET}'")
if academic_results:
    plot_model_performance(academic_results, f"Pr√©diction de '{ACADEMIC_TARGET}'")

# %% [markdown]
# ## √âtape 8 : S√©lection des 3 meilleurs mod√®les

# %%
def select_top_models(results_dict, metric='f1', n_top=3):
    """
    S√©lectionne les n meilleurs mod√®les bas√©s sur une m√©trique
    """
    if not results_dict:
        print("‚ùå Aucun r√©sultat disponible pour la s√©lection")
        return []

    # Calculer les scores pour la m√©trique choisie
    scores = {name: results[metric] for name, results in results_dict.items()}

    # Trier par score d√©croissant
    sorted_models = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:n_top]

    print(f"\nüèÜ TOP {n_top} MOD√àLES (bas√© sur {metric.upper()}):")
    for i, (name, score) in enumerate(sorted_models, 1):
        acc = results_dict[name]['accuracy']
        print(f"  {i}. {name}:")
        print(f"     {metric.upper()}: {score:.4f} | Accuracy: {acc:.4f}")

    return [name for name, _ in sorted_models]

print("="*60)
print("S√âLECTION DES MEILLEURS MOD√àLES")
print("="*60)

# S√©lection pour la t√¢che 1
if stress_results:
    print(f"\nüìä Pour la t√¢che 1 (Pr√©diction de '{STRESS_TARGET}'):")
    top_stress_models = select_top_models(stress_results, metric='f1', n_top=3)
else:
    print(f"\n‚ùå Pas de r√©sultats pour la t√¢che 1")
    top_stress_models = []

# S√©lection pour la t√¢che 2
if academic_results:
    print(f"\nüìä Pour la t√¢che 2 (Pr√©diction de '{ACADEMIC_TARGET}'):")
    top_academic_models = select_top_models(academic_results, metric='f1', n_top=3)
else:
    print(f"\nüìä M√™mes mod√®les pour la t√¢che 2 (m√™me target ou pas de r√©sultats)")
    top_academic_models = top_stress_models.copy()

# Visualisation comparative SANS SEABORN
def plot_top_models_comparison(stress_top, academic_top, stress_results, academic_results):
    """
    Comparaison visuelle des top mod√®les pour les deux t√¢ches
    """
    if not stress_top:
        print("‚ùå Aucun mod√®le √† comparer")
        return

    fig, axes = plt.subplots(1, 2 if academic_results else 1, figsize=(14, 6))

    if academic_results:
        ax1, ax2 = axes
    else:
        ax1 = axes
        ax2 = None

    # T√¢che 1
    stress_metrics = ['accuracy', 'f1']
    x = np.arange(len(stress_top))
    width = 0.35

    for idx, metric in enumerate(stress_metrics):
        offset = width * idx
        values = [stress_results[model][metric] for model in stress_top]
        ax1.bar(x + offset, values, width, label=metric.upper(),
               color=['skyblue', 'lightgreen'][idx], edgecolor='black')

    ax1.set_title(f"Top mod√®les - Pr√©diction de '{STRESS_TARGET}'", fontsize=12, fontweight='bold')
    ax1.set_xticks(x + width/2)
    ax1.set_xticklabels([m[:15] + '...' if len(m) > 15 else m for m in stress_top], rotation=45)
    ax1.set_ylabel('Score')
    ax1.legend()
    ax1.grid(axis='y', alpha=0.3)

    # T√¢che 2 (si diff√©rente)
    if ax2 is not None and academic_top:
        anxiety_metrics = ['accuracy', 'f1']
        x = np.arange(len(academic_top))

        for idx, metric in enumerate(anxiety_metrics):
            offset = width * idx
            values = [academic_results[model][metric] for model in academic_top]
            ax2.bar(x + offset, values, width, label=metric.upper(),
                   color=['skyblue', 'lightgreen'][idx], edgecolor='black')

        ax2.set_title(f"Top mod√®les - Pr√©diction de '{ACADEMIC_TARGET}'", fontsize=12, fontweight='bold')
        ax2.set_xticks(x + width/2)
        ax2.set_xticklabels([m[:15] + '...' if len(m) > 15 else m for m in academic_top], rotation=45)
        ax2.legend()
        ax2.grid(axis='y', alpha=0.3)

    plt.suptitle('Comparaison des top 3 mod√®les par t√¢che', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

if stress_results:
    plot_top_models_comparison(top_stress_models, top_academic_models, stress_results, academic_results)

# %% [markdown]
# ## √âtape 9 : Optimisation des hyperparam√®tres

# %%
def optimize_top_models(top_models, results_dict, X_train, y_train, X_val, y_val, preprocessor, task_name):
    """
    Optimise les hyperparam√®tres des meilleurs mod√®les
    """
    print(f"\n{'='*60}")
    print(f"OPTIMISATION DES HYPERPARAM√àTRES - {task_name}")
    print(f"{'='*60}")

    if not top_models:
        print("‚ùå Aucun mod√®le √† optimiser")
        return {}

    optimized_models = {}

    # Grilles d'hyperparam√®tres simplifi√©es pour √©viter les temps d'ex√©cution trop longs
    param_grids = {
        'Random Forest': {
            'n_estimators': [100, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        },
        'Gradient Boosting': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 4]
        },
        'XGBoost': {
            'n_estimators': [100, 200],
            'learning_rate': [0.01, 0.1],
            'max_depth': [3, 4]
        },
        'Logistic Regression': {
            'C': [0.1, 1, 10],
            'penalty': ['l2']
        },
        'SVM': {
            'C': [0.1, 1, 10],
            'kernel': ['linear', 'rbf']
        },
        'Decision Tree': {
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        },
        'K-Nearest Neighbors': {
            'n_neighbors': [3, 5, 7],
            'weights': ['uniform', 'distance']
        }
    }

    for model_name in top_models:
        if model_name in param_grids:
            print(f"\nüîß Optimisation de {model_name}...")

            # R√©cup√©rer le mod√®le de base
            if model_name in results_dict:
                base_model = results_dict[model_name]['model']
            else:
                print(f"  ‚ö†Ô∏è  Mod√®le {model_name} non trouv√© dans les r√©sultats, cr√©ation d'un nouveau")
                # Recr√©er le mod√®le
                model_creators = {
                    'Random Forest': RandomForestClassifier(random_state=42),
                    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
                    'XGBoost': XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
                    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
                    'SVM': SVC(probability=True, random_state=42),
                    'Decision Tree': DecisionTreeClassifier(random_state=42),
                    'K-Nearest Neighbors': KNeighborsClassifier()
                }
                base_model = model_creators.get(model_name, LogisticRegression(max_iter=1000, random_state=42))

            # Cr√©er un pipeline avec le preprocessor
            try:
                pipeline = Pipeline([
                    ('preprocessor', preprocessor),
                    ('classifier', base_model)
                ])

                # GridSearchCV avec param√®tres r√©duits pour plus de vitesse
                grid_search = GridSearchCV(
                    pipeline,
                    {f'classifier__{k}': v for k, v in param_grids[model_name].items()},
                    cv=3,  # R√©duit pour plus de vitesse
                    scoring='f1_weighted',
                    n_jobs=-1,
                    verbose=0
                )

                grid_search.fit(X_train, y_train)

                print(f"  ‚úÖ Meilleurs param√®tres trouv√©s")
                print(f"  üìä Meilleur score (CV): {grid_search.best_score_:.4f}")

                # √âvaluation sur validation
                y_pred = grid_search.predict(X_val)
                val_f1 = f1_score(y_val, y_pred, average='weighted', zero_division=0)
                print(f"  üìà F1 sur validation: {val_f1:.4f}")

                # Comparaison avec le mod√®le non optimis√© (si disponible)
                if model_name in results_dict:
                    original_f1 = results_dict[model_name]['f1']
                    improvement = val_f1 - original_f1
                    print(f"  üìà Am√©lioration: {improvement:+.4f}")

                optimized_models[model_name] = grid_search.best_estimator_

            except Exception as e:
                print(f"  ‚ùå Erreur lors de l'optimisation: {str(e)[:100]}...")
                print(f"  ‚ö†Ô∏è  Utilisation du mod√®le non optimis√©")
                # Create a pipeline with the base model, even if optimization failed
                pipeline = Pipeline([
                    ('preprocessor', preprocessor),
                    ('classifier', base_model)
                ])
                optimized_models[model_name] = pipeline

        else:
            print(f"\n‚ö†Ô∏è  Pas de grille d'hyperparam√®tres d√©finie pour {model_name}")
            print(f"  Utilisation du mod√®le tel quel")
            if model_name in results_dict:
                pipeline = Pipeline([
                    ('preprocessor', preprocessor),
                    ('classifier', results_dict[model_name]['model'])
                ])
            else:
                # Cr√©er un mod√®le par d√©faut
                pipeline = Pipeline([
                    ('preprocessor', preprocessor),
                    ('classifier', LogisticRegression(max_iter=1000, random_state=42))
                ])
            optimized_models[model_name] = pipeline

    return optimized_models

# Optimiser les mod√®les pour la t√¢che 1
if top_stress_models:
    print("\nüéØ OPTIMISATION POUR LA PR√âDICTION DU STRESS")
    optimized_stress_models = optimize_top_models(
        top_stress_models, stress_results,
        X_train_stress, y_train_stress, X_val_stress, y_val_stress,
        preprocessor_stress, f"Pr√©diction de '{STRESS_TARGET}'"
    )
else:
    print("\n‚ùå Pas de mod√®les √† optimiser pour le stress")
    optimized_stress_models = {}

# Optimiser les mod√®les pour la t√¢che 2 (Performance acad√©mique)
if top_academic_models:
    print("\nüéØ OPTIMISATION POUR LA PR√âDICTION DE LA PERFORMANCE ACAD√âMIQUE")
    optimized_academic_models = optimize_top_models(
        top_academic_models, academic_results,
        X_train_academic, y_train_academic, X_val_academic, y_val_academic,
        preprocessor_academic, f"Pr√©diction de '{ACADEMIC_TARGET}'"
    )
else:
    print("\n‚ùå Pas de mod√®les √† optimiser pour la performance acad√©mique")
    optimized_academic_models = {}

# %% [markdown]
# ## √âtape 10 : √âvaluation finale sur l'ensemble de test

# %%
def evaluate_on_test_set(optimized_models, X_test, y_test, task_name, label_encoder):
    """
    √âvalue les mod√®les optimis√©s sur l'ensemble de test
    """
    print(f"\n{'='*60}")
    print(f"√âVALUATION FINALE SUR TEST - {task_name}")
    print(f"{'='*60}")

    if not optimized_models:
        print("‚ùå Aucun mod√®le √† √©valuer")
        return {}

    final_results = {}

    for model_name, model in optimized_models.items():
        print(f"\nüìä √âvaluation de {model_name}:")

        try:
            # Pr√©dictions
            y_pred = model.predict(X_test)
            y_pred_proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None

            # M√©triques
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
            recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
            f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

            # AUC-ROC
            if y_pred_proba is not None:
                if len(label_encoder.classes_) == 2:
                    auc = roc_auc_score(y_test, y_pred_proba[:, 1])
                else:
                    auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr', average='weighted')
            else:
                auc = None

            # Matrice de confusion
            cm = confusion_matrix(y_test, y_pred)

            # Stocker les r√©sultats
            final_results[model_name] = {
                'accuracy': accuracy,
                'precision': precision,
                'recall': recall,
                'f1': f1,
                'auc': auc,
                'confusion_matrix': cm,
                'y_pred': y_pred
            }

            print(f"  ‚úÖ Accuracy: {accuracy:.4f}")
            print(f"  ‚úÖ Precision: {precision:.4f}")
            print(f"  ‚úÖ Recall: {recall:.4f}")
            print(f"  ‚úÖ F1-Score: {f1:.4f}")
            if auc:
                print(f"  ‚úÖ AUC-ROC: {auc:.4f}")

            # Taux d'erreur
            error_rate = 1 - accuracy
            print(f"  ‚ùå Taux d'erreur: {error_rate:.2%}")

            # Rapport de classification d√©taill√©
            print(f"\n  üìã Rapport de classification:")
            print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

            # Visualisation de la matrice de confusion
            plt.figure(figsize=(8, 6))
            disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
            disp.plot(cmap='Blues', values_format='d')
            plt.title(f'Matrice de confusion - {model_name}\n{task_name}', fontsize=12, fontweight='bold')
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"  ‚ùå Erreur lors de l'√©valuation: {str(e)[:100]}...")
            continue

    return final_results

# √âvaluation finale pour la t√¢che 1
if optimized_stress_models:
    print("\nüéØ √âVALUATION FINALE - PR√âDICTION DU STRESS")
    final_stress_results = evaluate_on_test_set(
        optimized_stress_models, X_test_stress, y_test_stress,
        f"Pr√©diction de '{STRESS_TARGET}'", le_stress
    )
else:
    print("\n‚ùå Pas de mod√®les optimis√©s pour le stress √† √©valuer")
    final_stress_results = {}

# √âvaluation finale pour la t√¢che 2 (Performance acad√©mique)
if optimized_academic_models and ACADEMIC_TARGET != STRESS_TARGET:
    print("\nüéØ √âVALUATION FINALE - PR√âDICTION DE LA PERFORMANCE ACAD√âMIQUE")
    final_academic_results = evaluate_on_test_set(
        optimized_academic_models, X_test_academic, y_test_academic,
        f"Pr√©diction de '{ACADEMIC_TARGET}'", le_academic
    )
else:
    print("\n‚ùå Pas de mod√®les optimis√©s pour la performance acad√©mique √† √©valuer")
    final_academic_results = {}

# %% [markdown]
# ## √âtape 11 : Analyse des erreurs et recommandations

# %%
def analyze_errors_and_generate_recommendations(final_results, X_test, y_test, task_name, target_column, label_encoder):
    """
    Analyse les erreurs et g√©n√®re des recommandations
    """
    print(f"\n{'='*60}")
    print(f"ANALYSE DES ERREURS - {task_name}")
    print(f"{'='*60}")

    if not final_results:
        print("‚ùå Aucun r√©sultat √† analyser")
        return None, 0

    # Trouver le meilleur mod√®le
    try:
        best_model_name = max(final_results.items(), key=lambda x: x[1]['f1'])[0]
        print(f"\nüèÜ Meilleur mod√®le: {best_model_name}")
        print(f"üìä F1-Score: {final_results[best_model_name]['f1']:.4f}")
        print(f"üéØ Accuracy: {final_results[best_model_name]['accuracy']:.4f}")
    except:
        print("‚ùå Impossible de d√©terminer le meilleur mod√®le")
        return None, 0

    # Analyser les erreurs
    # R√©cup√©rer le mod√®le optimis√© correspondant
    if 'stress' in task_name.lower():
        model_dict = optimized_stress_models
    else:
        model_dict = optimized_academic_models

    if best_model_name in model_dict:
        best_model = model_dict[best_model_name]
        y_pred = best_model.predict(X_test)
        errors_mask = y_test != y_pred

        if sum(errors_mask) > 0:
            # Make a copy of X_test to avoid SettingWithCopyWarning
            errors_df = X_test.loc[errors_mask].copy()
            errors_df['true_label'] = y_test[errors_mask]
            errors_df['predicted_label'] = y_pred[errors_mask]

            n_errors = len(errors_df)
            total_samples = len(X_test)
            error_rate = n_errors / total_samples

            print(f"\n‚ùå Analyse des erreurs:")
            print(f"  Nombre d'erreurs: {n_errors}/{total_samples} ({error_rate:.1%})")

            # Distribution des erreurs par type
            if n_errors > 0:
                print(f"\n  üìä Distribution des erreurs (vrai ‚Üí pr√©dit):")
                errors_df['true_label_decoded'] = label_encoder.inverse_transform(errors_df['true_label'])
                errors_df['predicted_label_decoded'] = label_encoder.inverse_transform(errors_df['predicted_label'])
                error_counts = errors_df.groupby(['true_label_decoded', 'predicted_label_decoded']).size().reset_index(name='count')
                for _, row in error_counts.iterrows():
                    print(f"    {row['true_label_decoded']} ‚Üí {row['predicted_label_decoded']}: {row['count']} erreurs")

            # Analyser les caract√©ristiques des erreurs pour les variables num√©riques
            numeric_cols = X_test.select_dtypes(include=['int64', 'float64']).columns

            if len(numeric_cols) > 0 and n_errors > 0:
                print(f"\n  üìà Analyse des caract√©ristiques num√©riques:")

                # Limiter √† 3 variables pour √©viter trop de graphiques
                numeric_cols = list(numeric_cols)[:3]

                fig, axes = plt.subplots(1, min(3, len(numeric_cols)), figsize=(15, 5))
                if len(numeric_cols) == 1:
                    axes = [axes]

                for idx, col in enumerate(numeric_cols):
                    if col in X_test.columns:
                        correct_data = X_test[~errors_mask][col]
                        error_data = errors_df[col]

                        # Boxplot comparatif
                        if not correct_data.empty and not error_data.empty:
                            bp = axes[idx].boxplot([correct_data.dropna(), error_data.dropna()],
                                                   labels=['Correct', 'Erreurs'],
                                                   patch_artist=True)

                            # Colorier les boxplots
                            bp['boxes'][0].set_facecolor('lightgreen')
                            bp['boxes'][1].set_facecolor('lightcoral')
                            bp['boxes'][0].set_edgecolor('black')
                            bp['boxes'][1].set_edgecolor('black')

                            axes[idx].set_title(f'{col}\nComparaison correct vs erreurs', fontsize=10)
                            axes[idx].set_ylabel(col)
                            axes[idx].grid(axis='y', alpha=0.3)

                            # Statistiques
                            correct_mean = correct_data.mean()
                            error_mean = error_data.mean()
                            print(f"    {col}:")
                            print(f"      Moyenne (correct): {correct_mean:.2f}")
                            print(f"      Moyenne (erreurs): {error_mean:.2f}")
                            print(f"      Diff√©rence: {error_mean - correct_mean:+.2f}")

                plt.suptitle(f'Caract√©ristiques des erreurs - {task_name}', fontsize=12, fontweight='bold')
                plt.tight_layout()
                plt.show()
    else:
        print(f"‚ùå Mod√®le {best_model_name} non trouv√© dans les mod√®les optimis√©s")
        error_rate = 1 - final_results[best_model_name]['accuracy']

    # G√©n√©rer des recommandations
    print(f"\nüí° RECOMMANDATIONS POUR AM√âLIORER LA PR√âDICTION:")
    print(f"  1. Collecter plus de donn√©es pour les classes sous-repr√©sent√©es")
    print(f"  2. Enrichir les features avec des donn√©es suppl√©mentaires")
    print(f"  3. Essayer des techniques d'ensemble (voting, stacking)")
    print(f"  4. Utiliser l'augmentation de donn√©es pour les classes minoritaires")

    return best_model_name, error_rate if 'error_rate' in locals() else 0

if final_stress_results:
    print("\nüîç ANALYSE DES ERREURS POUR LE STRESS")
    best_stress_model, stress_error_rate = analyze_errors_and_generate_recommendations(
        final_stress_results, X_test_stress, y_test_stress,
        f"Pr√©diction de '{STRESS_TARGET}'", STRESS_TARGET, le_stress
    )
else:
    print("\n‚ùå Pas de r√©sultats pour l'analyse des erreurs du stress")
    best_stress_model, stress_error_rate = None, 0

if final_academic_results and ACADEMIC_TARGET != STRESS_TARGET:
    print("\nüîç ANALYSE DES ERREURS POUR LA PERFORMANCE ACAD√âMIQUE")
    best_academic_model, academic_error_rate = analyze_errors_and_generate_recommendations(
        final_academic_results, X_test_academic, y_test_academic,
        f"Pr√©diction de '{ACADEMIC_TARGET}'", ACADEMIC_TARGET, le_academic
    )
else:
    print("\n‚ùå Pas de r√©sultats pour l'analyse des erreurs de la performance acad√©mique")
    best_academic_model, academic_error_rate = None, 0

# %% [markdown]
# ## √âtape 12 : Rapport final et visualisation

# %%
print("="*80)
print("RAPPORT FINAL D'ANALYSE - SYNTH√àSE COMPL√àTE")
print("="*80)

print(f"\nüìÅ DONN√âES ANALYS√âES:")
print(f"  ‚Ä¢ Fichier: {file_path}")
print(f"  ‚Ä¢ √âchantillons totaux: {len(df_clean)}")
print(f"  ‚Ä¢ Variables: {len(df_clean.columns)}")

print(f"\nüéØ VARIABLES CIBLES ANALYS√âES:")
print(f"  ‚Ä¢ T√¢che 1 (Stress): {STRESS_TARGET}")
print(f"  ‚Ä¢ T√¢che 2 (Performance acad√©mique): {ACADEMIC_TARGET}")

print(f"\nüìä PERFORMANCE DES MOD√àLES OPTIMIS√âS:")
if final_stress_results:
    print(f"\n  A) PR√âDICTION DE '{STRESS_TARGET}':")
    for model_name, metrics in final_stress_results.items():
        print(f"     ‚Ä¢ {model_name}:")
        print(f"       - F1-Score: {metrics['f1']:.4f}")
        print(f"       - Accuracy: {metrics['accuracy']:.4f}")
        print(f"       - Taux d'erreur: {(1-metrics['accuracy']):.2%}")
        if best_stress_model and model_name == best_stress_model:
            print(f"       - Meilleur mod√®le: ‚úì")
else:
    print(f"\n  A) PR√âDICTION DE '{STRESS_TARGET}': Aucun r√©sultat")

if final_academic_results and ACADEMIC_TARGET != STRESS_TARGET:
    print(f"\n  B) PR√âDICTION DE '{ACADEMIC_TARGET}':")
    for model_name, metrics in final_academic_results.items():
        print(f"     ‚Ä¢ {model_name}:")
        print(f"       - F1-Score: {metrics['f1']:.4f}")
        print(f"       - Accuracy: {metrics['accuracy']:.4f}")
        print(f"       - Taux d'erreur: {(1-metrics['accuracy']):.2%}")
        if best_academic_model and model_name == best_academic_model:
            print(f"       - Meilleur mod√®le: ‚úì")
elif ACADEMIC_TARGET != STRESS_TARGET:
    print(f"\n  B) PR√âDICTION DE '{ACADEMIC_TARGET}': Aucun r√©sultat")

print(f"\nüîç MARGES D'ERREUR ESTIM√âES:")
print(f"  ‚Ä¢ Pr√©diction du stress: {stress_error_rate:.2%}")
if ACADEMIC_TARGET != STRESS_TARGET:
    print(f"  ‚Ä¢ Pr√©diction de la performance acad√©mique: {academic_error_rate:.2%}")

print(f"\nüí° RECOMMANDATIONS POUR LA R√âDUCTION DU STRESS ET DE L'ANXI√âT√â:")

# Analyse bas√©e sur les donn√©es
print(f"\n  BAS√â SUR L'ANALYSE DES DONN√âES:")

# Si nous avons les colonnes attendues
key_metrics = ['Screen Time (hrs/day)', 'Sleep Duration (hrs)', 'Physical Activity (hrs/week)']

for metric in key_metrics:
    if metric in df_clean.columns:
        if STRESS_TARGET in df_clean.columns:
            # Analyser par niveau de stress
            stress_levels = df_clean[STRESS_TARGET].unique()
            print(f"\n  üìä {metric} par niveau de stress:")
            for level in sorted(stress_levels):
                subset = df_clean[df_clean[STRESS_TARGET] == level]
                if not subset.empty:
                    mean_val = subset[metric].mean()
                    print(f"     ‚Ä¢ {level}: {mean_val:.1f}")

print(f"\n  üéØ RECOMMANDATIONS CONCR√àTES:")
recommendations = """
    1. GESTION DU TEMPS D'√âCRAN:
       ‚Ä¢ Limiter √† moins de 6h/jour pendant les p√©riodes acad√©miques
       ‚Ä¢ Utiliser des applications de suivi du temps d'√©cran
       ‚Ä¢ Privil√©gier les cours en direct aux enregistrements

    2. AM√âLIORATION DU SOMMEIL:
       ‚Ä¢ Objectif: 7-8 heures de sommeil par nuit
       ‚Ä¢ √âtablir une routine de coucher r√©guli√®re
       ‚Ä¢ √âviter les √©crans 1 heure avant le sommeil

    3. ACTIVIT√â PHYSIQUE R√âGULI√àRE:
       ‚Ä¢ Minimum 150 minutes/semaine d'activit√© mod√©r√©e
       ‚Ä¢ Int√©grer des pauses actives pendant les sessions d'√©tude

    4. STRAT√âGIES D'APPRENTISSAGE:
       ‚Ä¢ Technique Pomodoro (25 min √©tude, 5 min pause)
       ‚Ä¢ R√©vision espac√©e plut√¥t que bachotage
       ‚Ä¢ √âtude en groupe pour r√©duire l'isolement

    5. SUPPORT √âMOTIONNEL:
       ‚Ä¢ Groupes de discussion entre peers
       ‚Ä¢ Acc√®s √† des conseillers acad√©miques
       ‚Ä¢ Techniques de pleine conscience
"""

print(recommendations)

print(f"\nüìà VISUALISATION FINALE DES R√âSULTATS")

# Cr√©er un dashboard final
try:
    # 1. Performance finale des mod√®les (Stress)
    if final_stress_results:
        stress_models = list(final_stress_results.keys())
        stress_f1 = [final_stress_results[m]['f1'] for m in stress_models]
        stress_acc = [final_stress_results[m]['accuracy'] for m in stress_models]

        x = np.arange(len(stress_models))
        width = 0.35

        fig, axes = plt.subplots(1, 2 if final_academic_results and ACADEMIC_TARGET != STRESS_TARGET else 1, figsize=(14, 6))
        if final_academic_results and ACADEMIC_TARGET != STRESS_TARGET:
            ax1, ax2 = axes
        else:
            ax1 = axes

        ax1.bar(x - width/2, stress_f1, width, label='F1-Score', color='skyblue', edgecolor='black')
        ax1.bar(x + width/2, stress_acc, width, label='Accuracy', color='lightgreen', edgecolor='black')
        ax1.set_title(f'Performance finale - {STRESS_TARGET}', fontsize=12, fontweight='bold')
        ax1.set_xticks(x)
        ax1.set_xticklabels([m[:10] + '...' if len(m) > 10 else m for m in stress_models], rotation=45)
        ax1.set_ylabel('Score')
        ax1.legend()
        ax1.grid(axis='y', alpha=0.3)

        # 2. Performance finale des mod√®les (Performance acad√©mique)
        if final_academic_results and ACADEMIC_TARGET != STRESS_TARGET:
            academic_models = list(final_academic_results.keys())
            academic_f1 = [final_academic_results[m]['f1'] for m in academic_models]
            academic_acc = [final_academic_results[m]['accuracy'] for m in academic_models]

            x = np.arange(len(academic_models))
            ax2.bar(x - width/2, academic_f1, width, label='F1-Score', color='skyblue', edgecolor='black')
            ax2.bar(x + width/2, academic_acc, width, label='Accuracy', color='lightgreen', edgecolor='black')
            ax2.set_title(f'Performance finale - {ACADEMIC_TARGET}', fontsize=12, fontweight='bold')
            ax2.set_xticks(x)
            ax2.set_xticklabels([m[:10] + '...' if len(m) > 10 else m for m in academic_models], rotation=45)
            ax2.legend()
            ax2.grid(axis='y', alpha=0.3)

        plt.suptitle('DASHBOARD FINAL - ANALYSE DE LA SANT√â MENTALE', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
except Exception as e:
    print(f"‚ö†Ô∏è  Erreur lors de la cr√©ation du dashboard: {str(e)[:100]}...")

print(f"\n‚úÖ ANALYSE TERMIN√âE AVEC SUCC√àS !")
print(f"\nüìÅ PROCHAINES √âTAPES:")
print(f"  1. Sauvegarder les mod√®les avec joblib ou pickle")
print(f"  2. D√©ployer l'API de pr√©diction avec Flask ou FastAPI")
print(f"  3. Mettre en place un monitoring continu des performances")
print(f"  4. G√©n√©rer des rapports r√©guliers pour les administrateurs")

# %% [markdown]
# ## √âtape 13 : Sauvegarde des mod√®les (optionnel)

# %%
# Sauvegarde des mod√®les si d√©sir√©
save_models = False  # Changez √† True si vous voulez sauvegarder

if save_models:
    print("=== SAUVEGARDE DES MOD√àLES ===")

    try:
        import joblib
        import os
        from datetime import datetime

        # Cr√©er le dossier si n√©cessaire
        os.makedirs('saved_models', exist_ok=True)

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

        if optimized_stress_models and best_stress_model:
            stress_model_path = f'saved_models/stress_model_{timestamp}.pkl'
            joblib.dump(optimized_stress_models[best_stress_model], stress_model_path)
            print(f"‚úÖ Mod√®le de stress sauvegard√©: {stress_model_path}")

        if optimized_academic_models and best_academic_model and ACADEMIC_TARGET != STRESS_TARGET:
            academic_model_path = f'saved_models/academic_model_{timestamp}.pkl'
            joblib.dump(optimized_academic_models[best_academic_model], academic_model_path)
            print(f"‚úÖ Mod√®le de performance acad√©mique sauvegard√©: {academic_model_path}")

        print(f"\nüìÅ Les mod√®les sont sauvegard√©s dans le dossier 'saved_models/'")

    except ImportError:
        print("‚ö†Ô∏è  Joblib n'est pas install√©. Installation recommand√©e: pip install joblib")
    except Exception as e:
        print(f"‚ùå Erreur lors de la sauvegarde: {str(e)[:100]}...")
else:
    print("‚ÑπÔ∏è  Pour sauvegarder les mod√®les, changez 'save_models = True' dans la cellule pr√©c√©dente")

# %% [markdown]
# ## √âtape 14 : Sauvegarde des 3 meilleurs mod√®les

# %%
print("="*80)
print("SAUVEGARDE DES 3 MEILLEURS MOD√àLES")
print("="*80)

try:
    import joblib
    import os
    from datetime import datetime
    import json

    # Cr√©er le dossier si n√©cessaire
    os.makedirs('saved_models', exist_ok=True)
    os.makedirs('saved_models/metadata', exist_ok=True)

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    # 1. Sauvegarde pour la t√¢che 1 (Stress)
    if final_stress_results and best_stress_model:
        print(f"\nüìÅ SAUVEGARDE DES MOD√àLES POUR LA T√ÇCHE 1: '{STRESS_TARGET}'")

        # Trier les mod√®les par F1-score
        stress_sorted = sorted(final_stress_results.items(),
                              key=lambda x: x[1]['f1'],
                              reverse=True)

        # Sauvegarder les 3 meilleurs mod√®les
        for i, (model_name, metrics) in enumerate(stress_sorted[:3], 1):
            if model_name in optimized_stress_models:
                model = optimized_stress_models[model_name]
                model_path = f'saved_models/stress_model_{i}_{timestamp}.pkl'

                # Sauvegarde du mod√®le
                joblib.dump(model, model_path)

                # Sauvegarde des m√©triques
                metrics_path = f'saved_models/metadata/stress_metrics_{i}_{timestamp}.json'
                with open(metrics_path, 'w') as f:
                    json.dump({
                        'model_name': model_name,
                        'target': STRESS_TARGET,
                        'timestamp': timestamp,
                        'metrics': {
                            'f1_score': float(metrics['f1']),
                            'accuracy': float(metrics['accuracy']),
                            'precision': float(metrics['precision']),
                            'recall': float(metrics['recall']),
                            'error_rate': float(1 - metrics['accuracy'])
                        },
                        'features': list(X_train_stress.columns)
                    }, f, indent=2)

                print(f"  ‚úÖ Mod√®le {i}: {model_name}")
                print(f"     F1-Score: {metrics['f1']:.4f}")
                print(f"     Accuracy: {metrics['accuracy']:.4f}")
                print(f"     Chemin: {model_path}")

    # 2. Sauvegarde pour la t√¢che 2 (Anxi√©t√©) si diff√©rente
    if 'final_academic_results' in locals() and final_academic_results and best_academic_model and ACADEMIC_TARGET != STRESS_TARGET: # Changed ANXIETY_TARGET to ACADEMIC_TARGET
        print(f"\nüìÅ SAUVEGARDE DES MOD√àLES POUR LA T√ÇCHE 2: '{ACADEMIC_TARGET}'")

        # Trier les mod√®les par F1-score
        academic_sorted = sorted(final_academic_results.items(), # Changed anxiety_sorted to academic_sorted
                               key=lambda x: x[1]['f1'],
                               reverse=True)

        # Sauvegarder les 3 meilleurs mod√®les
        for i, (model_name, metrics) in enumerate(academic_sorted[:3], 1):
            if model_name in optimized_academic_models: # Changed optimized_anxiety_models to optimized_academic_models
                model = optimized_academic_models[model_name] # Changed optimized_anxiety_models to optimized_academic_models
                model_path = f'saved_models/academic_model_{i}_{timestamp}.pkl' # Changed anxiety_model to academic_model

                # Sauvegarde du mod√®le
                joblib.dump(model, model_path)

                # Sauvegarde des m√©triques
                metrics_path = f'saved_models/metadata/academic_metrics_{i}_{timestamp}.json' # Changed anxiety_metrics to academic_metrics
                with open(metrics_path, 'w') as f:
                    json.dump({
                        'model_name': model_name,
                        'target': ACADEMIC_TARGET, # Changed ANXIETY_TARGET to ACADEMIC_TARGET
                        'timestamp': timestamp,
                        'metrics': {
                            'f1_score': float(metrics['f1']),
                            'accuracy': float(metrics['accuracy']),
                            'precision': float(metrics['precision']),
                            'recall': float(metrics['recall']),
                            'error_rate': float(1 - metrics['accuracy'])
                        },
                        'features': list(X_train_academic.columns) # Changed X_train_anxiety to X_train_academic
                    }, f, indent=2)

                print(f"  ‚úÖ Mod√®le {i}: {model_name}")
                print(f"     F1-Score: {metrics['f1']:.4f}")
                print(f"     Accuracy: {metrics['accuracy']:.4f}")
                print(f"     Chemin: {model_path}")

    # 3. Sauvegarde des donn√©es de configuration
    print(f"\nüìä SAUVEGARDE DES CONFIGURATIONS ET M√âTADONN√âES")

    config = {
        'timestamp': timestamp,
        'dataset_info': {
            'file_path': file_path,
            'samples': len(df_clean),
            'features': len(df_clean.columns),
            'targets': {
                'stress_target': STRESS_TARGET,
                'academic_target': ACADEMIC_TARGET # Changed anxiety_target to academic_target
            }
        },
        'preprocessing_info': {
            'missing_values_handled': bool(missing_after > 0), # Ensure native Python bool
            'categorical_features_processed': len(categorical_cols)
        },
        'best_models': {
            'stress': best_stress_model,
            'academic': best_academic_model if ACADEMIC_TARGET != STRESS_TARGET else 'same_as_stress' # Changed anxiety to academic
        }
    }

    config_path = f'saved_models/config_{timestamp}.json'
    with open(config_path, 'w') as f:
        json.dump(config, f, indent=2)

    print(f"  ‚úÖ Configuration sauvegard√©e: {config_path}")

    # 4. Cr√©er un fichier README avec les instructions
    readme_content = f"""
    # MOD√àLES SAUVEGARD√âS - Analyse de la Sant√© Mentale des √âtudiants
    Date: {timestamp}

    ## MOD√àLES DISPONIBLES

    ### 1. Pr√©diction du Stress ('{STRESS_TARGET}')
    """

    if final_stress_results and best_stress_model:
        stress_sorted = sorted(final_stress_results.items(), key=lambda x: x[1]['f1'], reverse=True)[:3]
        for i, (model_name, metrics) in enumerate(stress_sorted, 1):
            readme_content += f"""
    Mod√®le {i}: {model_name}
      - F1-Score: {metrics['f1']:.4f}
      - Accuracy: {metrics['accuracy']:.4f}
      - Fichier: stress_model_{i}_{timestamp}.pkl
      - M√©tadonn√©es: metadata/stress_metrics_{i}_{timestamp}.json
    """

    if 'final_academic_results' in locals() and final_academic_results and best_academic_model and ACADEMIC_TARGET != STRESS_TARGET: # Changed ANXIETY_TARGET to ACADEMIC_TARGET
        readme_content += f"""

    ### 2. Pr√©diction de la Performance Acad√©mique ('{ACADEMIC_TARGET}') # Changed Anxi√©t√© to Performance Acad√©mique
    """
        academic_sorted = sorted(final_academic_results.items(), key=lambda x: x[1]['f1'], reverse=True)[:3] # Changed anxiety_sorted to academic_sorted
        for i, (model_name, metrics) in enumerate(academic_sorted, 1):
            readme_content += f"""
    Mod√®le {i}: {model_name}
      - F1-Score: {metrics['f1']:.4f}
      - Accuracy: {metrics['accuracy']:.4f}
      - Fichier: academic_model_{i}_{timestamp}.pkl # Changed anxiety_model to academic_model
      - M√©tadonn√©es: metadata/academic_metrics_{i}_{timestamp}.json # Changed anxiety_metrics to academic_metrics
    """

    readme_content += f"""

    ## UTILISATION DES MOD√àLES

    ```python
    import joblib

    # Charger un mod√®le
    model = joblib.load('saved_models/stress_model_1_{timestamp}.pkl')

    # Faire une pr√©diction
    # Note: Les donn√©es doivent √™tre pr√©trait√©es comme pendant l'entra√Ænement
    prediction = model.predict(nouvelles_donn√©es)
    ```

    ## INFORMATIONS SUR LES DONN√âES
    - Dataset: {file_path}
    - √âchantillons: {len(df_clean)}
    - Variables: {len(df_clean.columns)}
    - Date de cr√©ation: {timestamp}
    """

    readme_path = f'saved_models/README_{timestamp}.md'
    with open(readme_path, 'w', encoding='utf-8') as f:
        f.write(readme_content)

    print(f"  ‚úÖ README cr√©√©: {readme_path}")

    print(f"\nüéâ SAUVEGARDE TERMIN√âE !")
    print(f"üìÇ Dossier: saved_models/")
    print(f"‚è∞ Timestamp: {timestamp}")

    # Afficher le r√©sum√©
    print(f"\nüìã R√âSUM√â DES MOD√àLES SAUVEGARD√âS:")

    # Compter les mod√®les sauvegard√©s
    import glob
    stress_models_count = len(glob.glob('saved_models/stress_model_*.pkl'))
    academic_models_count = len(glob.glob('saved_models/academic_model_*.pkl')) # Changed anxiety_models to academic_models

    print(f"  ‚Ä¢ Mod√®les de stress: {stress_models_count}")
    print(f"  ‚Ä¢ Mod√®les de performance acad√©mique: {academic_models_count}") # Changed d'anxi√©t√© to de performance acad√©mique
    print(f"  ‚Ä¢ Fichiers de m√©tadonn√©es: {len(glob.glob('saved_models/metadata/*.json'))}")

except ImportError as e:
    print(f"‚ùå Erreur d'importation: {e}")
    print("‚ö†Ô∏è  Installation requise: pip install joblib")

except Exception as e:
    print(f"‚ùå Erreur lors de la sauvegarde: {str(e)[:100]}...")
    print("‚ö†Ô∏è  V√©rifiez que vous avez les permissions d'√©criture dans le dossier courant.")
````

````python
def generate_suggestion(student_data, n_increase=1.0, k_decrease=1.0):
    """
    G√©n√®re des suggestions pour am√©liorer la sant√© mentale et la performance acad√©mique
    bas√©es sur les mod√®les entra√Æn√©s.

    Args:
        student_data (pd.Series): Profil de l'√©tudiant (une ligne du DataFrame).
        n_increase (float): Facteur d'augmentation pour le sommeil et l'activit√© physique.
        k_decrease (float): Facteur de diminution pour le temps d'√©cran.

    Returns:
        dict: Un dictionnaire contenant les pr√©dictions initiales, les pr√©dictions sugg√©r√©es,
              et les modifications des variables.
    """
    print("\n=== G√âN√âRATION DE SUGGESTIONS ===")

    # 1. R√©cup√©rer les meilleurs mod√®les et encodeurs
    try:
        best_stress_pipeline = optimized_stress_models[best_stress_model]
        best_academic_pipeline = optimized_academic_models[best_academic_model]
        stress_le = le_stress
        academic_le = le_academic
    except NameError as e:
        print(f"\n‚ùå ERREUR: Variables de mod√®le ou d'encodeur non d√©finies. Ex√©cutez les √©tapes pr√©c√©dentes. {e}")
        return {"status": "error", "message": "Mod√®les ou encodeurs non charg√©s"}

    # Pr√©parer les donn√©es de l'√©tudiant pour la pr√©diction
    # Supprimer les colonnes cibles si elles sont pr√©sentes, puis convertir en DataFrame
    student_features = student_data.drop(
        columns=[STRESS_TARGET, ACADEMIC_TARGET], errors='ignore'
    ).to_frame().T

    # Assurer que l'ordre des colonnes est le m√™me que celui d'entra√Ænement
    # Utiliser les colonnes de X_train_stress comme r√©f√©rence (les deux t√¢ches ont les m√™mes features d'entr√©e)
    if 'X_train_stress' in globals():
        student_features = student_features[X_train_stress.columns]
    else:
        print("\n‚ùå ERREUR: X_train_stress non d√©fini. Ne peut pas assurer l'ordre des colonnes.")
        return {"status": "error", "message": "X_train_stress non d√©fini"}

    print(f"\nüîç Profil initial de l'√©tudiant:\n{student_features.iloc[0]}")

    # 2. et 3. Pr√©dictions initiales
    initial_stress_pred_encoded = best_stress_pipeline.predict(student_features)[0]
    initial_academic_pred_encoded = best_academic_pipeline.predict(student_features)[0]

    # 4. D√©coder les pr√©dictions
    initial_stress_pred_decoded = stress_le.inverse_transform([initial_stress_pred_encoded])[0]
    initial_academic_pred_decoded = academic_le.inverse_transform([initial_academic_pred_encoded])[0]

    # 5. Afficher les pr√©dictions initiales
    print(f"\nüéØ Pr√©diction initiale du Stress: {initial_stress_pred_decoded}")
    print(f"üéØ Pr√©diction initiale de la Performance Acad√©mique: {initial_academic_pred_decoded}")

    # Retourner les informations initiales pour la suite
    return {
        "student_features": student_features,
        "initial_stress_pred_decoded": initial_stress_pred_decoded,
        "initial_academic_pred_decoded": initial_academic_pred_decoded
    }

````

````python
import pandas as pd
import numpy as np

print("="*80)
print("√âTAPE 15 : FONCTION DE SUGGESTION PERSONNALIS√âE")
print("="*80)

# Assurez-vous que les meilleurs mod√®les et LabelEncoders sont disponibles
# (Ils devraient l'√™tre suite aux √©tapes pr√©c√©dentes)
if 'optimized_stress_models' not in locals() or 'best_stress_model' not in locals() or \
   'optimized_academic_models' not in locals() or 'best_academic_model' not in locals() or \
   'le_stress' not in locals() or 'le_academic' not in locals() or 'df_clean' not in locals():
    print("‚ùå Erreur: Les mod√®les ou les donn√©es n√©cessaires ne sont pas disponibles.")
    print("Veuillez r√©ex√©cuter les √©tapes pr√©c√©dentes pour charger et entra√Æner les mod√®les.")
    raise SystemExit("Arr√™t du programme. Mod√®les ou donn√©es manquants.")

# R√©cup√©rer les meilleurs pipelines (mod√®le + pr√©processeur)
best_stress_pipeline = optimized_stress_models[best_stress_model]
best_academic_pipeline = optimized_academic_models[best_academic_model]

def generate_suggestion(student_data, n_increase=1, k_decrease=1):
    """
    G√©n√®re une suggestion personnalis√©e pour un √©tudiant en simulant des changements
    dans le temps d'√©cran, le sommeil et l'activit√© physique, puis pr√©dit l'impact.

    Args:
        student_data (pd.Series): Une ligne du DataFrame `df_clean` repr√©sentant un √©tudiant.
        n_increase (float): Nombre d'heures √† augmenter pour le sommeil et l'activit√© physique.
        k_decrease (float): Nombre d'heures √† diminuer pour le temps d'√©cran.

    Returns:
        str: Une recommandation textuelle personnalis√©e.
    """
    if not isinstance(student_data, pd.Series):
        return "Erreur: Les donn√©es de l'√©tudiant doivent √™tre une pd.Series."
    if not all(col in student_data.index for col in ['Screen Time (hrs/day)', 'Sleep Duration (hrs)', 'Physical Activity (hrs/week)']):
        return "Erreur: Les colonnes d'activit√© cl√©s ('Screen Time (hrs/day)', 'Sleep Duration (hrs)', 'Physical Activity (hrs/week)') sont manquantes."

    print(f"\n--- G√©n√©ration de suggestion pour l'√©tudiant ---")
    print(f"Param√®tres de changement: +{n_increase}h sommeil/activit√©, -{k_decrease}h temps d'√©cran")

    # 1. Pr√©dictions initiales
    # Cr√©er un DataFrame √† partir de la Series pour la pr√©diction

    # Prepare features for stress prediction
    current_stress_features = pd.DataFrame([student_data.drop(columns=[STRESS_TARGET], errors='ignore')])
    current_stress_features = current_stress_features[X_train_stress.columns]

    # Prepare features for academic performance prediction
    current_academic_features = pd.DataFrame([student_data.drop(columns=[ACADEMIC_TARGET], errors='ignore')])
    current_academic_features = current_academic_features[X_train_academic.columns]

    initial_stress_pred_encoded = best_stress_pipeline.predict(current_stress_features)[0]
    initial_academic_pred_encoded = best_academic_pipeline.predict(current_academic_features)[0]

    initial_stress_level = le_stress.inverse_transform([initial_stress_pred_encoded])[0]
    initial_academic_perf = le_academic.inverse_transform([initial_academic_pred_encoded])[0]

    print(f"Pr√©dictions initiales: Stress Level = {initial_stress_level}, Academic Performance Change = {initial_academic_perf}")

    # 2. Simuler les changements
    hypothetical_student_data = student_data.copy()

    # Appliquer les changements et respecter les bornes
    hypothetical_student_data['Screen Time (hrs/day)'] = max(0, hypothetical_student_data['Screen Time (hrs/day)'] - k_decrease)
    hypothetical_student_data['Sleep Duration (hrs)'] = min(12, hypothetical_student_data['Sleep Duration (hrs)'] + n_increase) # Max 12h de sommeil
    hypothetical_student_data['Physical Activity (hrs/week)'] = min(20, hypothetical_student_data['Physical Activity (hrs/week)'] + n_increase) # Max 20h d'activit√©

    # Create DataFrames for hypothetical predictions
    hypothetical_stress_features = pd.DataFrame([hypothetical_student_data.drop(columns=[STRESS_TARGET], errors='ignore')])
    hypothetical_stress_features = hypothetical_stress_features[X_train_stress.columns]

    hypothetical_academic_features = pd.DataFrame([hypothetical_student_data.drop(columns=[ACADEMIC_TARGET], errors='ignore')])
    hypothetical_academic_features = hypothetical_academic_features[X_train_academic.columns]

    # 3. Pr√©dictions apr√®s simulation
    hypothetical_stress_pred_encoded = best_stress_pipeline.predict(hypothetical_stress_features)[0]
    hypothetical_academic_pred_encoded = best_academic_pipeline.predict(hypothetical_academic_features)[0]

    hypothetical_stress_level = le_stress.inverse_transform([hypothetical_stress_pred_encoded])[0]
    hypothetical_academic_perf = le_academic.inverse_transform([hypothetical_academic_pred_encoded])[0]

    print(f"Pr√©dictions simul√©es: Stress Level = {hypothetical_stress_level}, Academic Performance Change = {hypothetical_academic_perf}")

    # 4. Interpr√©ter et formuler la suggestion
    recommendation = f"Cher √©tudiant:\n\n"
    recommendation += f"Votre profil actuel sugg√®re un niveau de stress '{initial_stress_level}' et une performance acad√©mique '{initial_academic_perf}'.\n\n"
    recommendation += f"Si vous r√©duisiez votre temps d'√©cran de {k_decrease:.1f}h/jour et augmentiez votre dur√©e de sommeil et votre activit√© physique de {n_increase:.1f}h chacun:\n"
    recommendation += f"  - Votre temps d'√©cran passerait √† environ {hypothetical_student_data['Screen Time (hrs/day)']:.1f}h/jour.\n"
    recommendation += f"  - Votre dur√©e de sommeil passerait √† environ {hypothetical_student_data['Sleep Duration (hrs)']:.1f}h/nuit.\n"
    recommendation += f"  - Votre activit√© physique passerait √† environ {hypothetical_student_data['Physical Activity (hrs/week)']:.1f}h/semaine.\n\n"

    # Comparaison du stress
    stress_impact = ""
    if initial_stress_level == hypothetical_stress_level:
        stress_impact = f"Votre niveau de stress resterait probablement '{hypothetical_stress_level}'."
    else:
        stress_impact = f"Votre niveau de stress pourrait √©voluer de '{initial_stress_level}' √† '{hypothetical_stress_level}'."

    # Comparaison de la performance acad√©mique
    academic_impact = ""
    if initial_academic_perf == hypothetical_academic_perf:
        academic_impact = f"Votre performance acad√©mique resterait probablement '{hypothetical_academic_perf}'."
    else:
        academic_impact = f"Votre performance acad√©mique pourrait √©voluer de '{initial_academic_perf}' √† '{hypothetical_academic_perf}'."

    recommendation += f"Nos mod√®les pr√©voient que suite √† ces changements:\n"
    recommendation += f"  - {stress_impact}\n"
    recommendation += f"  - {academic_impact}\n\n"

    recommendation += f"Ces ajustements sont des pistes pour am√©liorer votre bien-√™tre et vos r√©sultats acad√©miques. "
    recommendation += f"Il est toujours recommand√© de consulter des professionnels de la sant√© pour des conseils personnalis√©s."

    return recommendation

# --- Exemple d'utilisation de la fonction de suggestion ---
print("\n" + "="*60)
print("D√âMONSTRATION DE LA FONCTION DE SUGGESTION")
print("="*60)

# S√©lectionner un √©tudiant exemple (par exemple, le premier √©tudiant du DataFrame nettoy√©)
example_student_id = 0
example_student = df_clean.iloc[example_student_id]

# D√©finir les param√®tres de changement
n_increase_example = 4.0 # Augmenter le sommeil et l'activit√© physique de 1.5 heures
k_decrease_example = 4.0 # Diminuer le temps d'√©cran de 2.0 heures

suggestion_text = generate_suggestion(example_student, n_increase_example, k_decrease_example)
print("\n" + suggestion_text)

# Un autre exemple avec un √©tudiant diff√©rent et des param√®tres diff√©rents
example_student_id_2 = 100
example_student_2 = df_clean.iloc[example_student_id_2]
n_increase_example_2 = 3
k_decrease_example_2 = 4

suggestion_text_2 = generate_suggestion(example_student_2, n_increase_example_2, k_decrease_example_2)
print("\n" + suggestion_text_2)

print("\n" + "="*80)
print("R√âSUM√â DE LA FONCTIONNALIT√â AJOUT√âE")
print("="*80)
print("La nouvelle fonction `generate_suggestion` permet de cr√©er des recommandations personnalis√©es pour les √©tudiants.")
print("En simulant des modifications dans des facteurs cl√©s comme le temps d'√©cran, la dur√©e de sommeil et l'activit√© physique,")
print("elle pr√©dit l'impact sur les niveaux de stress et la performance acad√©mique.")
print("Cette fonctionnalit√© offre un outil puissant pour:")
print("  - **Personnaliser les conseils**: Adapter les recommandations aux profils individuels des √©tudiants.")
print("  - **Visualiser l'impact**: Montrer concr√®tement comment de petits changements de comportement peuvent influencer leur bien-√™tre et leurs r√©sultats.")
print("  - **Promouvoir des habitudes saines**: Encourager les √©tudiants √† adopter des modes de vie plus √©quilibr√©s gr√¢ce √† des preuves bas√©es sur les mod√®les.")
print("Cette √©tape transforme nos mod√®les pr√©dictifs en un syst√®me de soutien proactif, aidant les √©tudiants √† prendre des d√©cisions √©clair√©es pour leur sant√© mentale et leur succ√®s acad√©mique.")
````
