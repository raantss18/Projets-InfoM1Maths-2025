```python
#PROJET 10 : SÉRIE DE FOURIER
import numpy as np
# Calcul du coefficient a_0
def a_0():
    return 0  
# Calcul des coefficients a_n (tous nuls ici)
def a_n(n):
    return 0  # Car x * cos(n*x) est une fonction impaire
# Calcul des coefficients b_n
def b_n(n):
    return (2 * (-1)**(n+1)) / n
# Nombre de coefficients à calculer
N = 10 
# Calcul des coefficients
a0 = a_0()
a_coeffs = [a_n(n) for n in range(1, N+1)]
b_coeffs = [b_n(n) for n in range(1, N+1)]
print(f"a_0 = {a0}")
for n in range(1, N+1):
    print(f"a_{n} = {a_coeffs[n-1]:.1f} , b_{n} = {b_coeffs[n-1]:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# f(x) = x
def f(x):
    return x

# Définition  S_N(x)
def S_N(x, N):
    sum_terms = np.zeros_like(x)
    for n in range(1, N+1):
        sum_terms += (2 * (-1)**(n+1) / n) * np.sin(n * x)
    return sum_terms
x = np.linspace(-np.pi, np.pi, 400)
# Choisir valeur de N
N_values = [5, 10, 20, 50]  
plt.figure(figsize=(10, 6))
for N in N_values:
    plt.plot(x, S_N(x, N), linestyle="dashed", label=f"S_N(x) avec N={N}")

plt.plot(x, f(x), label="f(x) = x", color="black", linewidth=2)
# Lignes pour x = ±π pour visualiser Gibbs
plt.axvline(-np.pi, color='gray', linestyle='dotted', linewidth=0.8)
plt.axvline(np.pi, color='gray', linestyle='dotted', linewidth=0.8)

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

plt.legend()
plt.title("Approximation de f(x) = x par sa série de Fourier et phénomène de Gibbs")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Affichage du graphique
plt.show()
```

INTERPRÉTATION
- Pour N est petit , l'approximation S_N(x) est très éloignée de f(x) .
- Pour N est grand , la série de Fourier dévient plus précise et suit mieux la courbe de f(x) sur [-pi , pi]
- Si S_N(x) converge ponctuellement vers f(x) , il ne converge pas uniformément sur tout l'intérvalle
